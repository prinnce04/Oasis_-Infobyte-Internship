```python
figure = plt.figure(figsize= (10,3))
sns.boxplot(df['price'],orient= "h")
plt.title("price")
plt.show()
```

```python
# finding IQR, Q1, Q3, min for Year_Birth

q1 = df['price'].quantile(0.25)
q3 = df['price'].quantile(0.75)

IQR = q3 - q1
min_ = q1 - 1.5*IQR

print("Q1 = ", q1,"\n"
      "Q3 = ", q3,"\n"
      "IQR = ", IQR,"\n"
      "Minumum = ", min_)
```

```python
out_index = df[df['price'] <= 1000].index  # fetching the indexes of outlying rows.
out_index
```

```python
df.loc[out_index,"price"]
```

```python
df.loc[out_index, "price"] = min_               #replacing the value with minimum
```
